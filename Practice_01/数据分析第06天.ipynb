```python
persons_df = pd.DataFrame(
    data={
    '姓名': ['关羽', '张飞', '赵云', '马超', '黄忠'],
    '职业': ['医生', '医生', '程序员', '画家', '教师'],
    '学历': ['研究生', '大专', '研究生', '高中', '本科'],
    }
)
persons_df
```

```python
persons_df.drop(columns=['职业'], inplace=True) # 删除职业列
```

```python
persons_df
```

```python
def edu_to_num(edu):
    if edu == '研究生':
        return 10
    elif edu == '本科':
        return 5
    elif edu == '大专':
        return 3
    elif edu == '高中':
        return 1
    else:
        return 0

persons_df['学历'] = persons_df.学历.apply(edu_to_num) # 将学历转换为数字
```

**分组聚合**

```python
sales_df['月份'] = sales_df.销售日期.dt.month # 提取月份
sales_df.groupby('月份')[['销售额']].sum() # 按月份分组，计算销售额
```

```python
ser = sales_df.groupby('品牌').销售额.sum()
ser
```

```python
ser.plot(kind='pie', autopct='%.1f%%', pctdistance=1.3) # 绘制柱状图
```

```python
temp_df = sales_df.groupby(['销售区域', '月份'])[['销售额']].sum() # 按销售区域和月份分组，计算销售额
temp_df 
```

```python
temp_df = temp_df.reset_index() # 重置索引
```

```python
temp_df.pivot(index='销售区域', columns='月份', values='销售额').fillna(0).applymap(int) # 将数据透视为表格
```

```python
sales_df.pivot_table(index='销售区域', columns='月份', values='销售额', aggfunc='sum',
                     fill_value=0, margins=True, margins_name='总计').applymap(int)  # 将数据透视为表格,并计算总计,并将空值填充为0,并将数据转换为整数
```

```python
sales_df.info()
```

```python
sales_df.pivot_table(index='销售渠道', columns='品牌', values='销售数量',
                     aggfunc='sum', margins=True, margins_name='总计')  # 将数据透视为表格,并计算总计
```

```python
def price(x):
    if x < 200:
        return '低端'
    elif 200 <= x < 500:
        return '中端'
    else:
        return '高端'
    
sales_df['价位'] = sales_df.售价.apply(price)
sales_df
```

```python
sales_df['销售占比'] = sales_df.销售数量 / sales_df.销售数量.sum()
sales_df
```

```python
sales_df.pivot_table(index='价位', columns='月份', values='销售占比', aggfunc='sum', margins=True, margins_name='总计').applymap(lambda x: '---' if np.isnan(x) else f'{x * 100:.2f}%') # 将数据透视为表格,并计算总计,并将空值填充为0,并将数据转换为百分比
```

```python
seg = pd.cut(sales_df.售价, bins=[0, 200 ,470 ,1500], right=True, labels=['低端', '中端', '高端']) # 将销售额分段
seg
```

```python
sales_df.pivot_table(index=seg, columns='月份', values='销售占比',
                     aggfunc='sum')  # 将数据透视为表格,并计算总计
```

```python
seg = pd.qcut(sales_df.售价, q=[0, 0.25, 0.5, 0.75], labels=['低端', '中端', '高端']) # 将销售额分段
seg
```

```python
sales_df.pivot_table(index=seg, columns='月份', values='销售占比',
                     aggfunc='sum')  # 将数据透视为表格,并计算总计
```
