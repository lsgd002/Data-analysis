```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_iris
```

```python
# 加载数据集
datasets = load_iris()
```

```python
# 查看数据集的属性
dir(datasets)
```

```python
print(datasets.DESCR)
```

```python
# 获取特征数据
feature_data = datasets.data
feature_data
```

```python
type(feature_data)
```

```python
feature_data.shape
```

```python
# 获取特征名称
feature_names = datasets.feature_names
feature_names
```

```python
# 获取花萼长度
sepal_length = feature_data[:, 0]
sepal_length
```

```python
# 获取花瓣长度
petal_length = feature_data[:, 2]
petal_length
```

```python
# 花萼长度和花瓣长度的关系
np.corrcoef(sepal_length, petal_length)
```

```python
# 绘制散点图
plt.scatter(sepal_length, petal_length)
plt.show()
```

```python
# 获取花瓣宽度
petal_width = feature_data[:, 3]
petal_width
```

```python
# 花瓣长度和花瓣宽度的关系
np.corrcoef(petal_length, petal_width)
```

```python
# 绘制散点图
plt.scatter(petal_length, petal_width)
plt.show()
```

```python
# least squares - 最小二乘解 --> 四元组（斜率和截距，误差平方和）
results = np.linalg.lstsq(feature_data[:, 2:4], feature_data[:, 0], rcond=None)
results
```

```python
# 获取斜率和截距（降到1维）
a, b = results[0].flatten()
a, b
```

```python
# 多项式最小二乘拟合
np.polyfit(feature_data[:, 2], feature_data[:, 0], 1)
```

```python
# 绘制散点图
plt.scatter(feature_data[:, 2], feature_data[:, 0])
# 绘制拟合直线
plt.plot(feature_data[:, 2], a * feature_data[:, 2] + b, 'r')
plt.show()
```
